## Código criado para facilitar a leitura dos títulos e resumos

```python
import json
import pandas as pd

docs = {}

with open("data/corpus.jsonl", "r", encoding="utf-8") as f:
    for line in f:
        d = json.loads(line)
        doc_id = d.get("arxiv_id") or d.get("id")

        if doc_id:
            docs[str(doc_id).strip()] = d

pool = pd.read_csv(
    "eval/qrels_pool.tsv",
    sep=r"\s+",
    usecols=[0,1,2],
    names=["qid", "iter", "doc_id"]
)

for _, row in pool.iterrows():
    doc_id = str(row["doc_id"]).strip()
    doc = docs.get(doc_id)

    print("=" * 80)
    print("QID:", row["qid"])
    print("DOC:", doc_id)

    if doc is None:
        print("DOCUMENTO NÃO ENCONTRADO NO CORPUS")
        continue

    print("TITLE:", doc.get("title", ""))
    print()
    print(doc.get("abstract", ""))
    print()
```
